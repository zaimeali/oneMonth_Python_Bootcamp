# import Pandas

```python
import pandas as pd
```

# import Data

```python
header = ['id', 'title', 'year', 'rating', 'votes', 'length', 'genres']
# index_col = 0 will make the first row as id
data = pd.read_csv('imdb_top_10000.txt', sep = '\t', names = header, index_col = 0)
```

# Explore Data

```python
data.head(10)
```

```python
data.info()
```

```python
data.describe()
```

```python
data.describe(include='all')
```

# exporting Data

```python
data.to_csv('test.csv', header=True, index=True, sep=",")
```

# Sorting Data

```python
data.sort_values(by = 'rating')
```

```python
data.sort_values(by = 'year')
```

```python
# in descending order
data.sort_values(by = 'rating', ascending = False)
```
